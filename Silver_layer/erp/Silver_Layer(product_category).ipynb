### Checking Dupliactes

```python
from pyspark.sql.functions import *
```

```python
df.groupBy("ID") \
  .count() \
  .filter("count > 1") \
  .display()
```

### Handling Null Values 

```python
from pyspark.sql.functions import *

df.select(
    *[
        sum(col(c).isNull().cast("int")).alias(c)
        for c in df.columns
    ]
).display()
```

#### Changing Column names

```python
df.columns
```

```python
rename_map = {
    "ID": "category_id",
    "CAT": "category",
    "SUBCAT": "subcategory"
}
```

```python
for old, new in rename_map.items():
    df = df.withColumnRenamed(old, new)

df.columns
```

```python
df.display()
```

```python
# 10. Write Silver
df.write \
    .format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable("bike_data.silver.product_category_details")
```
