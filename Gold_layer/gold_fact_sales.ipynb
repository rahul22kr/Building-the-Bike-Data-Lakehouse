#### fact table

```python
%sql
WITH base AS (
    SELECT
        f.ord_num,
        f.prd_key,
        f.customer_id,
        f.order_dt,
        f.ship_dt,
        f.due_dt,
        f.quantity,
        f.price,
        f.sales
    FROM bike_data.silver.fact_sales_details f
)

SELECT DISTINCT *
FROM base;
```

```python
df_sales = spark.sql("""WITH base AS (
    SELECT
        f.ord_num,
        f.prd_key,
        f.customer_id,
        f.order_dt,
        f.ship_dt,
        f.due_dt,
        f.quantity,
        f.price,
        f.sales
    FROM bike_data.silver.fact_sales_details f
)

SELECT DISTINCT *
FROM base;""")
```

```python
df_sales.display()
```

#### checking Duplicate

```python
df_sales.groupBy("ord_num", "prd_key").count().filter("count > 1").show()
```

#### writing data

```python
df_sales.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bike_data.gold.fact_sales")
```
