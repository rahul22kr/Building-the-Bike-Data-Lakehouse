#### Joining all relevant Silver product tables 

```python
df_products=spark.sql("""WITH base AS (
    SELECT
        p.product_id,
        p.product_key,
        p.product_nm,
        p.product_cost,
        p.product_line,
        p.product_start_dt,
        p.product_end_dt,
        c.category_id,
        c.category,
        c.subcategory,
        c.MAINTENANCE
    FROM bike_data.silver.product_details p
    LEFT JOIN bike_data.silver.product_category_details c
        ON p.category_id = c.category_id
)

SELECT DISTINCT *
FROM base
""")
```

```python
df_products.display()
```

#### writing data

```python
df_products.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bike_data.gold.dim_products")
```
