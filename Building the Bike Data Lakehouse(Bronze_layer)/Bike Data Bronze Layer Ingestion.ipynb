### Creating Catalog

```python
spark.sql("""
          create catalog if not exists Bike_data
          """)
```

### Creaing Schema

```python
spark.sql("""
CREATE SCHEMA IF NOT EXISTS Bike_data.bronze
""")
```

### # Read multiple raw CSV files from a Unity Catalog volume into Spark DataFrames with header parsing and automatic schema inference for Bronze ingestion.


```python
cust_az12 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/CUST_AZ12.csv")

loc_a101 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/LOC_A101.csv")

px_cat_g1v2 = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/PX_CAT_G1V2.csv")

cust_info = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/cust_info.csv")

prd_info = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/prd_info.csv")

sales_details = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/raw_data/default/raw_data_bike/sales_details.csv")
```

### Writing the  DataFrames to a Bronze Delta table in overwrite mode 


```python
cust_info.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.cust_info_crm")
```

```python
cust_az12.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.cust_az12_erp")
```

```python
loc_a101.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.loc_a101_erp")
```

```python
px_cat_g1v2.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.px_cat_g1v2_erp")
    
```

```python
prd_info.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.prd_info_crm")
```

```python
cust_info.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.cust_info_crm")
```

```python
sales_details.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bike_data.bronze.sales_details_crm")
```
